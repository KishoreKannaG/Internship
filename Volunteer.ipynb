```python
def vol_count(state_name):
  nyv_count = df.loc[df.index == state_name]['National Youth Volunteers'].sum()
  print(nyv_count)

for state in df.index:
  print(state, end=" ")
  vol_count(state)
```

```python
def male_vol_count(state_name):
  male_count = df.loc[df.index == state_name]['Male Member'].sum()
  print(male_count)

for state in df.index:
  print(state, end=" ")
  male_vol_count(state)
```

```python
def female_vol_count(state_name):
  female_count = df.loc[df.index == state_name]['Female Member'].sum()
  print(female_count)

for state in df.index:
  print(state, end=" ")
  female_vol_count(state)
```

#Data Visualization

statewise volunteers distribution by catogary

```python
def state_chart(state_name):
  state_filter = df.index == state_name
  filtered_df = df[state_filter]
  category_counts = filtered_df[['General', 'SC', 'ST', 'OBC']].sum()
  plt.figure(figsize=(4, 4))
  plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
  plt.title('Distribution of Categories in '+ str(state_name))
  plt.axis('equal')
  plt.show()

for state in df.index:
  state_chart(state)
```

statewise gender participition ratio

```python
def gender_chart(state_name):
  state_filter = df.index == state_name
  filtered_df = df[state_filter]
  male_count = filtered_df['Male Member'].sum()
  female_count = filtered_df['Female Member'].sum()
  plt.figure(figsize=(5, 5))
  plt.bar(['Male', 'Female'], [male_count, female_count])
  plt.title('Distribution of Male and Female Members in '+str(state_name))
  plt.xlabel('Gender')
  plt.ylabel('Count')
  plt.show()

for state in df.index:
  gender_chart(state)
```

overall contribution by category

```python
category_counts = df[['General', 'SC', 'ST', 'OBC']].sum()
category_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Categories')
plt.show()
```

male and female comparision

```python
plt.figure(figsize=(13,5))
x = plt.xticks(rotation=90)
sns.lineplot(data=df['Male Member'])
sns.lineplot(data=df['Female Member'])
```

commulative bar graph

```python
def catagory(id):
  plt.figure(figsize=(16, 7))
  x = plt.xticks(rotation=90)
  sns.barplot(x=df.index, y=df[id])
  plt.show()

for id in df.columns:
  catagory(id)
```

heatmap for correlation of each field

```python
columns = [ 'National Youth Volunteers', 'Male Member', 'Female Member', 'General', 'SC', 'ST', 'OBC', '10th', '12th', 'Graduate', 'Post Graduate']
selected_df = df[columns]
correlation_matrix = selected_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
